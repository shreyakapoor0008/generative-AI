```python
#import library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/data/placement (1).csv")
df
```

```python
# find number of rows and column
df.shape
```

```python
#find the information of the data
df.info()
```

```python
#find the null value in the data set
df.isnull().sum()
```

```python
#find independent and dependent feature
x= df.iloc[:,0]
y= df.iloc[:,1]
x
```

```python
#spilt the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

x_train
```

```python
#training the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train.values.reshape(-1,1),y_train)
```

```python
#coefficient
m = model.coef_
print(m[0])
```

```python
#intercept
b = model.intercept_
print(b)
```

```python
cgpa = 9.5
package = m*cgpa + b
print(package)

```

```python
#testing the model
y_pred = model.predict(x_test.values.reshape(-1,1))
y_pred, y_test
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["cgpa"],df["package"])
plt.plot(x_test, y_pred, color="red")
plt.xlabel("CGPA")
plt.ylabel("PACKAGE")
plt.show()


```
